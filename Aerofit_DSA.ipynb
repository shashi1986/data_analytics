```python
cat=sns.countplot(data=df,x='Product')
for bar in cat.patches:
    # Get the height of the bar (which represents the count)
    count = int(bar.get_height())
    # Position the text (x, y) slightly above the bar
    cat.text(
        bar.get_x() + bar.get_width() / 2,  # Horizontal center of the bar
        bar.get_height() + 0.5,            # Slightly above the bar
        str(count),                        # Count as text
        ha='center', fontsize=10          # Text alignment and font size
    )
plt.title("Sales by Product")
plt.show()

#Insight 3: KP281 are the highest selling product.
```

```python
cat=sns.countplot(data=df,x='Product',hue='Gender')
for bar in cat.patches:
    # Get the height of the bar (which represents the count)
    count = int(bar.get_height())
    # Position the text (x, y) slightly above the bar
    cat.text(
        bar.get_x() + bar.get_width() / 2,  # Horizontal center of the bar
        bar.get_height() + 0.5,            # Slightly above the bar
        str(count),                        # Count as text
        ha='center', fontsize=10          # Text alignment and font size
    )
plt.title("Sales by Product by Gender")
plt.show()

#Insight 4 : Amongst KP781 Male customers are more  ~ 82 %
```

```python
l_b= df['Income'].quantile(0.25)
u_b= df['Income'].quantile(0.75)
iqr=u_b-l_b
outlier=u_b+1.5*iqr
print(outlier)
df[df['Income']>69000]['Product'].value_counts()
#Insight 7:  Income > 69000 has bought only KP781 . KP781 can be marketed to Higher income group to get greater sales.
```

```python
l_b= df['Miles'].quantile(0.25)
u_b= df['Miles'].quantile(0.75)
iqr=u_b-l_b
outlier=u_b+1.5*iqr
print(outlier)
df[df['Miles']>outlier]['Product'].value_counts()
#Insight 7:  People who did more miles  bought more KP781 . KP781 can be marketed to Higher miles group to get greater sales.
```
